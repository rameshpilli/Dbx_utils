```
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql import SQLContext
sc = spark.sparkContext
```

```
schema = StructType([
    StructField("ID", IntegerType(), True)
    ,StructField("Version", FloatType(), True)
    ,StructField("Details", StringType(), True)
])

df = spark.createDataFrame(sc.emptyRDD(),schema)
df.show()
```

```
#add details into the empty df 
df = spark.createDataFrame([
    [1, 0.1, """{"name": "abc", 
                "age": 15,
                "occupation": "student"}"""]
    ,[2, 0.1, """{"name": "xyz",
                  "age": 45,
                  "occupation": "entrepreneur"}"""]
    ,[3, 0.1, """{"name": "ghi", 
                    "age": 50,
                    "occupation": "doctor"}"""]
], schema)
display(df)
```

```
#write to schema
df.write.mode("overwrite").saveAsTable("default.temp_table")
```

```
import pandas as pd

def search_column(columnName):
    lst = []
    for db in spark.sql('''show databases''').rdd.collect():
        for row in spark.sql(f'''show tables from {db['databaseName']}''').rdd.collect():
            try:
                for column in spark.sql('''show columns from '''+ row['database'] + '''.''' + row['tableName']).rdd.collect():
                    if columnName.lower() in column['col_name'].lower():
                        lst.append([row['database'] + '''.''' + row['tableName'], column['col_name']])
            except:
                pass
            
    df = pd.DataFrame(lst, columns=['TableName', 'ColumnName'])
    return df

df = search_column("ID")
display(df)
```
